Before starting the assignment let us import the dataset:

```r
cholesterol_data<-read.table("cholesterol.txt",header=TRUE)
str(cholesterol_data) #looking at the table
```


a) Make some relevant plots of this data set, comment on normality. Investigate whether the columns Before and After8weeks are correlated.

To check if the columns are correlated we can check the the scatter plot  to check if it has a linear realtion:

```r

plot(cholesterol_data$Before, cholesterol_data$After8weeks)

# Calculate the correlation coefficient
correlation_value <- cor(cholesterol_data$Before, cholesterol_data$After8weeks)

# Print the correlation value
print(paste("Correlation:", correlation_value))
```

As we can see from the correlation value that is 0.99 ≈1 and the graph, we can assume there is linear relationship between the 2 variable Before and After8Weeks

To check the normality we can check the QQ plot and the Shapiro-Wilk test, after checking the normality of both we also check the normality of the difference bewteen the before and after the use of margarine.

```r


qqnorm(cholesterol_data$Before)  # normal QQ-plot
qqline(cholesterol_data$Before)
shapiro.test(cholesterol_data$Before)  #Shapiro-Wilk test

qqnorm(cholesterol_data$After8weeks)
qqline(cholesterol_data$After8weeks)
shapiro.test(cholesterol_data$After8weeks)
```
